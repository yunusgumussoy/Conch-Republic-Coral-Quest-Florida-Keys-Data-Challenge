```python
# Handle fractional hours properly
CREMP_Temperatures_2023["Timestamp"] = pd.to_datetime(
    CREMP_Temperatures_2023[["Year", "Month", "Day"]]
) + pd.to_timedelta(CREMP_Temperatures_2023["Time"], unit="h")
```

```python
CREMP_Temperatures_2023.head()
```

```python
# Optional: Convert Timestamp to date if you want daily avg matching
CREMP_Temperatures_2023["Date"] = CREMP_Temperatures_2023["Timestamp"].dt.date
```

```python
CREMP_Temperatures_2023.head()
```

```python
CREMP_Temperatures_2023.info()
```

```python
CREMP_Temperatures_2023["Date"] = pd.to_datetime(CREMP_Temperatures_2023["Timestamp"]).dt.normalize()
```

```python
CREMP_Temperatures_2023.info()
```

```python
# Merge (example: average temperature per site per day)
temp_daily_avg = CREMP_Temperatures_2023.groupby(["SiteID", "Date"])["TempC"].mean().reset_index()

scor_data = scor_data.merge(
    temp_daily_avg,
    on=["SiteID", "Date"],
    how="left"
)
```

```python
scor_data.info()
```

```python
# Save data
scor_data.to_csv("scor_data.csv", index=False)
```
